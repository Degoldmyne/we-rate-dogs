Now, suppose we ask Twitter to help us out. Twitter points us to a handle named @dog_rates. The screen name of the handle is WeRateDogs. Dog owners send pictures of their dogs to this handle and the dogs are rated. We gat some data for a few years on the breed of dogs that has been sent for rating. What do we find?

Well check out the top 7 dog breeds according to we rate dogs (earlier than 2017 tho :)):

```python
twitter_archive_master = pd.read_csv('twitter_archive_master.csv')
breeds = list(twitter_archive_master['dog_breed'].value_counts().index[:7])
twitter_archive_master = twitter_archive_master.loc[twitter_archive_master['dog_breed'].isin(breeds)]

twitter_archive_master.dog_breed.value_counts().plot(kind='bar');
plt.title('Chart Showing Top 7 Dog Breeds', fontsize=18, pad=5)
plt.xlabel('Dog Breeds', fontsize=18)
plt.ylabel('Counts', fontsize=18)
plt.show()
```

```python
twitter_archive_master.dog_breed.value_counts().plot(kind='pie');
plt.title('Pie Chart Showing Top Seven Dog breeds');
```

Alright. We got one more question for you:
    
If a random person is swipping through some random dog tweets on WeRateDogs, which dog breed is most likely to get more likes?

Well, the chart below gives us the numbers:

```python
#Let's plot the mean favorite count on a chart
mean_favorite_count = twitter_archive_master.groupby('dog_breed')['favorite_count'].mean()
mean_favorite_count.plot(kind='barh', title='Mean Favorite Count/Likes For Top 7 Dog Breeds', alpha=.2)
plt.xlabel('Dog Breeds')
plt.ylabel('Favorite_count/Likes')
plt.show();
```

Now, to our final question for you:
    
Which dog breed is most likely to score higher given a few picture of dog breeds in the USA?

The chart below definitely confirms you are correct

```python
mean_score_rating = twitter_archive_master.groupby(['dog_breed'])['rating_numerator'].mean()
mean_score_rating.plot(kind='bar', title='Mean Rating Score For Top 7 Dog Breeds', alpha=.2)
plt.xlabel('Dog Breeds')
plt.ylabel('Mean rating Score')
plt.show();
```

The bars seem to be telling the same stories. The Golden Retriever and Labrador Retriever seem to be the people's favorite in the United States.
